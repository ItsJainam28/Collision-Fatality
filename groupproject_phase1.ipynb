```python
from sklearn.impute import SimpleImputer
features = ['ROAD_CLASS', 'DISTRICT', 'LOCCOORD', 'TRAFFCTL', 'VISIBILITY', 'LIGHT', 'RDSFCOND', 'IMPACTYPE', 'INVAGE', 'INVTYPE', 'PEDESTRIAN', 'CYCLIST', 'AUTOMOBILE', 'MOTORCYCLE', 'TRUCK','TRSN_CITY_VEH','PASSERGERS', 'SPEEDING', 'AG_DRIV', 'ALCOHOL', 'NEIGHBOURHOOD_158' , 'INJURY' ]

def remove_unwanted_columns(df):
    global features
    features_list = list(df.columns.values)
    for column in features_list:
        if column not in features:
            df = df.drop(column, axis=1)
    return df

def prepare_injury(df):
    classes = pd.value_counts(df['INJURY'])
    blank = classes.index[2]
    classes = classes.drop(blank)
    df = modify_injury(df, blank)
    return df

def modify_injury(df, blank):
    df['INJURY'].replace(to_replace="None", value=-1, inplace=True)
    df['INJURY'].replace(to_replace="Minor", value=-1, inplace=True)
    df['INJURY'].replace(to_replace="Minimal", value=-1, inplace=True)
    df['INJURY'].replace(to_replace="Major", value=+1, inplace=True)
    df['INJURY'].replace(to_replace="Fatal", value=+1, inplace=True)
    lst = df.index[df['INJURY'] == blank].tolist()
    df = df.drop(df.index[lst])
    df = df.reset_index(drop=True)
    return df

def handle_null_values(df):
    df = prepare_injury(df)
    df = df.replace(" ", np.nan)
    df = df.replace("", np.nan)
    no_columns = ['PEDESTRIAN', 'CYCLIST', 'AUTOMOBILE', 'MOTORCYCLE', 'TRUCK','TRSN_CITY_VEH','PASSENGERS', 'SPEEDING', 'AG_DRIV', 'ALCOHOL']
    for column in no_columns:
        df[column] = df[column].fillna('No')
    
    missing_columns = ['ROAD_CLASS', 'DISTRICT', 'LOCCOORD', 'TRAFFCTL', 'VISIBILITY', 'LIGHT', 'RDSFCOND', 'IMPACTYPE', 'INVAGE', 'INVTYPE', 'NEIGHBOURHOOD_158']

    # Define the imputer
    imputer = SimpleImputer(strategy='most_frequent')

    # Apply the imputer to the missing_columns
    df[missing_columns] = imputer.fit_transform(df[missing_columns])

    return df

df = remove_unwanted_columns(df)
df = handle_null_values(df)
df.isnull().sum()

    
```
