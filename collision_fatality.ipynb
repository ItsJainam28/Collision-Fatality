```python
import pandas as pd
```

```python
df = pd.read_csv('KSI.csv')
# How big is the dataset?
print(df.shape)
# Sample values
df.sample(5)
```

```python
# Mathematical representation of the data
df.info()
# Number of Null Values
df.isnull().sum()
# Corelation between the nymeric values
def is_fatal(injury):
    if injury == 'Fatal':
        return 1
    else:
        return 0

df['is_fatal'] = df['INJURY'].apply(is_fatal)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation = numeric_df.corr()['is_fatal']
print(correlation)
```

```python
# Categorical data with Unique values
categorical_columns = df.select_dtypes(include=['object']).columns
unique_counts = df[categorical_columns].nunique()
print(unique_counts)


```

```python
# Removing columns with high unique values
filtered_columns = unique_counts[unique_counts <= 160].index
columns_to_drop = ['HOOD_140', 'HOOD_158', 'NEIGHBOURHOOD_140', 'DIVISION', 'INJURY', 'ACCLASS', 'INITDIR', 'AUTOMOBILE']
filtered_columns = filtered_columns.drop(columns_to_drop, errors='ignore')
# Finiding the null values
null_counts = df[filtered_columns].isnull().sum()
print("Null values in filtered columns: ",null_counts)
df_encoded = pd.get_dummies(df[filtered_columns])
# Add 'IS_FATAL' column to encoded DataFrame
df_encoded['IS_FATAL'] = df['is_fatal']
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df_encoded[df_encoded.IS_FATAL==0]
df_minority = df_encoded[df_encoded.IS_FATAL==1]

# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=df_majority.shape[0],    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df = pd.concat([df_majority, df_minority_upsampled])

# Display new class counts
print(df.IS_FATAL.value_counts())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
# Assuming 'IS_FATAL' is your target variable
X = df.drop('IS_FATAL', axis=1)

y = df['IS_FATAL']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression(max_iter=100000, class_weight='balanced')

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)


# Print the accuracy of the model
print("Accuracy:", accuracy_score(y_test, y_pred))

# Print the confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Print the classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
